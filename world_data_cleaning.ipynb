# Global Factors Influencing Tertiary Education Enrollment
## Data cleaning

This data set contains information on world countries, values as at the year 2019. The objective is to understand what factors influence enrollment in tertiary education enrollment. We will first look at the data, generate a profile report, and clean and prepare it for analysis.

```python
import pandas as pd
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt

%matplotlib inline
```

We have 19 features (18 numeric and 1 categorical) describing 203 observations. We also see that there are missing values in some columns. Looking into them will help us determine whether we can fill the missing values or delete these observations. The only categorical column is the country name, which we could consider using as our identifier, though not necessary. This is also the only column with no missing values, meaning that there is no data for at least one country.

```python
df.describe()
```

Since our objective focuses on tertiary education enrollment, the target, looking briefly at descriptive statistics for this column, we see that the minimum enrollment is 0% and maximum is 148.53%. Typically, percentage scales up to 100%, so ordinarily this would appear to be an anomaly and warrant further investigation. However, in this case, column descriptions help us explain this. The inclusion of over- and under-aged students (due to repetition or early/late entrance) in a given level of education can cause enrollment to exceed 100%.

```python
df_missing = df[df.isna().any(axis=1)]
df_missing
```

"High income" and "Middle income" are not exactly country names, but represent groups of countries with similar characteristics, we could consider dropping them also. We can also check if any other income groups are listed in the dataset:

```python
df_other = df[df['Country Name'].str.contains('income')]
df_other
```

Other income levels are listed here. For our purpose, these observations could be helpful for comparison and overall evaluation. We can try filling the missing values for high- and middle-income before resorting to deletion. 

For High income, adolescent fertility rate and literacy rates are missing. Further research revealed that adolescent fertility rate for high income countries was 11.33 in 2019, according to the United States Federal Reserve Economic Data. Literacy data, however, is not readily available. This is the case for most of the missing data.

Before deleting the income groups with missing values, we can check whether countries within these groups are represented in this dataset:

Checking the dataset, there are individual high income countries listed. Therefore, removing the "High income" entry should not significantly affect our analysis since this income group is already represented.

```python
# https://ourworldindata.org/grapher/world-banks-income-groups
# middle income countries based on World Bank groupings
middle_income = ['China', 'Russia', 'Brazil', 'Mexico', 'India', 'Iran', 'Bolivia', 'Nicaragua']
mi_countries = df[df['Country Name'].isin(middle_income)]
mi_countries
```

```python
mi_2 = df[df['Country Name'].str.contains('Russia') | df['Country Name'].str.contains('Iran')]
mi_2
```

```python
df = df.drop([62, 117, 162, 167])
df.shape
```

```python
df = df.reset_index(drop=True)
```

No more missing values! 

Now, we have descriptive column headings, but they are a bit lengthy. We can shorten them to make them easier to read while keeping them informative of what each feature represents.

### renaming columns

```python
df.columns
```

```python
df.rename(columns = {'Country Name':'country_name', 'School enrollment, tertiary (% gross)':'tertiary_%', \
                     'School enrollment, secondary (% gross)':'secondary_%', \
                     'School enrollment, primary (% gross)':'primary_%', \
                     'Government expenditure on education, total (% of government expenditure)':'govt_exp_edu_%', \
                     'Literacy rate, adult total (% of people ages 15 and above)': 'adult_literacy', \
                     'Literacy rate, youth total (% of people ages 15-24)':'youth_literacy', \
                     'Birth rate, crude (per 1,000 people)':'birth_rate', \
                     'Mortality rate, infant (per 1,000 live births)':'infant_mortality_rate', \
                     'Life expectancy at birth, total (years)':'life_expectancy', \
                     'Adolescent fertility rate (births per 1,000 women ages 15-19)':'adolescent_fertility_rate', \
                     'Urban population (% of total population)':'urban_pop_%', \
                     'Access to electricity (% of population)':'electricity_access_%', \
                     'People using safely managed drinking water services (% of population)/at least basic': 'water_access_%', \
                     'Physicians (per 1,000 people)':'physicians_%',  \
                     'Unemployment, total (% of total labor force) (modeled ILO estimate)': 'unemployment_rate', \
                     'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)':'salaried_%', \
                     'Vulnerable employment, total (% of total employment) (modeled ILO estimate)':'vulnerable_%', \
                     'Inflation, consumer prices (annual %)':'inflation'}, inplace = True)
df.columns
```

```python
df.info()
```

```python
# converting values per 1000 to percentage (per 100)
df['birth_rate'] = df['birth_rate'].apply(lambda x: x / 10)
df['infant_mortality_rate'] = df['infant_mortality_rate'].apply(lambda x: x / 10)
df['adolescent_fertility_rate'] = df['adolescent_fertility_rate'].apply(lambda x: x / 10)
df['physicians_%'] = df['physicians_%'].apply(lambda x: x / 10)
```

### EDA

```python
sns.heatmap(df.phik_matrix())
```

It looks like among the features shown, tertiary education enrollment is least affected by primary school enrollment, inflation, government expenditure on education, and unemployment rate. Therefore, we will drop these columns.

```python
df_copy = df.copy()
df_copy = df_copy.drop(columns=['primary_%','inflation','govt_exp_edu_%','unemployment_rate'])
sns.heatmap(df_copy.phik_matrix())
```

```python
df_copy.shape
```

```python
# pearson correlation matrix for numeric columns
corrmat = df_copy.drop(columns=['country_name']).corr()
corrmat
```

Of the 15 features, secondary school enrollment, literacy, life expectancy, urban population, utility access, physician density and salaried employment percentage positively correlate to the target, which makes sense. Birth rate, infant mortality rate, adolescent fertility rate, and vulnerable employment percentage negatively correlate to the target.

```python
# strongest positive correlation to target
df.plot.scatter(x='tertiary_%', y='life_expectancy')
```

```python
# strongest negative correlation to target
df.plot.scatter(x='tertiary_%', y='birth_rate')
```

```python
import numpy as np

k = 14
cols = corrmat.nlargest(k, 'tertiary_%')['tertiary_%'].index

cm = np.corrcoef(df_copy[cols].values.T)
sns.set(font_scale = 1.25)
f, ax = plt.subplots(figsize=(10,8))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':12},\
                 yticklabels=cols.values, xticklabels=cols.values)
plt.show
```
